```python
# Step 1: Install Dependencies
!pip install -q openai-whisper gTTS
!sudo apt-get install -y ffmpeg

# Step 2: Upload Your Audio File (.mp3, .wav)
from google.colab import files
uploaded = files.upload()
audio_path = next(iter(uploaded))  # Use the first uploaded file

# Step 3: Transcribe Using Whisper (Forced CPU to avoid CUDA error)
import whisper
import torch

# Force model to run on CPU
device = "cpu"
model = whisper.load_model("base", device=device)
result = model.transcribe(audio_path)
input_text = result["text"]
print("🔊 Transcribed Text:", input_text)

# Step 4: Dialogue Logic (Simple Rules)
def reply(text):
    text = text.lower()

    # English Responses
    if "hello" in text or "hi" in text:
        return "Hello! How can I help you today?", "en"
    elif "booking" in text:
        return "Sure, I can help you with a booking. Please tell me the date and time.", "en"
    elif "where" in text or "location" in text or "office" in text:
        return "We are located in downtown Doha near Souq Waqif.", "en"
    elif "agent" in text or "human" in text or "talk to a person" in text:
        return "Please hold while I transfer you to a human agent...", "en"

    # Arabic Responses
    elif "أين" in text or "مكتب" in text:
        return "نحن موجودون في وسط الدوحة بالقرب من سوق واقف.", "ar"
    elif "حجز" in text:
        return "بالتأكيد، يمكنني مساعدتك في الحجز. من فضلك أخبرني بالتاريخ والوقت.", "ar"
    # Fallback Case (If Unrecognized Input)
    else:
        return "I'm sorry, I didn't understand that. Can you please repeat?", "en"

response_text, lang = reply(input_text)
print("🤖 AI Response:", response_text)

# Step 5: Convert AI Response to Speech
from gtts import gTTS
from IPython.display import Audio

# Generate the speech in the appropriate language (English or Arabic)
tts = gTTS(response_text, lang=lang)
tts.save("response.mp3")

# Step 6: Play AI Voice Response
Audio("response.mp3")
```
